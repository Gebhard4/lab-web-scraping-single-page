```python
import requests
import pandas as pd

from bs4 import BeautifulSoup 
```

### Scrape the current top 100 songs and their respective artists, and put the information into a pandas dataframe.

```python
r = requests.get('http://www.popvortex.com/music/charts/top-100-songs.php')
r.status_code
```

```python
html = r.content
html
```

```python
soup = BeautifulSoup(html, 'html.parser')
soup.head()
```

```python
charts = soup.find_all(class_="title")
charts
```

```python
song_names=[]
```

```python
for a in soup.find_all('cite',class_="title"):
    song_names.append(a.get_text())
```

```python
song_names
```

```python
artist = soup.find_all(class_="artist")
artist
```

```python
artists= []
```

```python
for a in soup.find_all('em',class_="artist"):
    artists.append(a.get_text())
```

```python
artists
```

```python
len(artists)
```

```python
len(song_names)
```

```python
df = pd.DataFrame(
    {"charts": song_names,
     "artists": artist})

df
```
